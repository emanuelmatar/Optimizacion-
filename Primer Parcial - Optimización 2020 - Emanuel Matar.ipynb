$OP-P1-E10$ Deseamos hallar en cada caso las funciones $f$ y los dominios $\Omega$ y los $x_*\in\Omega$ tales que:\
$(a)$ $x_*$ es un minimizador local de $f$ en $\Omega$ y $\nabla f(x_*)\neq 0$:\
Tomando $f(x,y)=x$, $\Omega=I^2$ donde $I=[0,1]$, y $x_*=(0,0)$.
Claramente $f(0,0)\leq f(x,y)$ $\forall (x,y)\in I^2$ pues $0=min${$x:x\in I$} pero $\nabla f(x,y)=(1,0)\neq(0,0)$


$(b)$ $x_*$ es un minimizador local de $f$ en $\Omega$ y $\nabla^2 f(x_*)$ no es semidefinida positiva:\
Tomamos ahora $f(x)=cos(x)$, $\Omega=[0,\pi/4]$, y $x_*=\pi/4$\
Como $f'(x)=-sen(x)$ es negativa en $(0,\pi/4]$, $cos(x)$ es decreciente en dicho intervalo, así $cos(\pi/4)=\frac{\sqrt{2}}{2} \leq cos(x)$ $\forall x\in [0,\pi/4]$. En este caso, el gradiente es la derivada respecto de la variable $x$ y vemos que vale $f'(x_*)=-sen(\pi/4)=-\frac{\sqrt{2}}{2} $ y la Hessiana es la derivada segunda respecto de $x$, $f''(x)=-cos(x)$ de donde:\
$\nabla^2 f(x_*)=-cos(\pi/4)=-\frac{\sqrt{2}}{2}<0$


$(c)$ $x_*\in \Omega$, $\Omega$ abierto, $\nabla f(x_*)= 0$ pero $x_*$ no es un minimizador local de $f$:\
En este caso acudimos a $f(x)=x^3$, $\Omega=\mathbb{R}$, y $x_*=0$.
Nuevamente, la derivada de $f$ ocupa el rol del gradiente:
$\nabla f(x)=f'(x)=3x^2$\
y al evaluarla en nuestro punto obtenemos:\
$\nabla f(x_*)=f'(0)=0$\
Ahora bien, $x^3$ es una función estrictamente creciente, y así:\
$x<0<y\Leftrightarrow x^3<0<y^3$,\
de donde concluimos que $0$ no es un minimizador local de la función en $\mathbb{R}$ pues no lo es en ningún entorno que contenga a $0$ por lo anteriormente observado. 


$(d)$ $x_*\in \Omega$, $\Omega$ abierto, $\nabla f(x_*)= 0$, $\nabla^2 f(x_*)\geq 0$ pero $x_*$ no es un minimizador local de $f$:\
$f(x)=x^{2n+1}$, $n\in\mathbb{N}$, $\Omega=\mathbb{R}$, y $x_*=0$\
Nuevamente, como el grado de la función es impar, estamos frente a una función estritamente creciente, y su derivada juega el rol de gradiente por ser de una variable. Calculamos entonces su derivada y la evaluamos:\
$f'(x)=(2n+1)x^{2n}$, $f'(0)=0$.\
Como es estrictamente creciente, $x<0<y\Leftrightarrow x^{2n+1}<0<y^{2n+1}$ y así, en todo entorno que contenga del punto $0$, el mismo no es un minimizador local.


$(e)$ $x_*\in \Omega$, $\Omega$ abierto, $x_*$ es un minimizador local de $f$ pero $\nabla^2 f(x_*)$ no es definida positiva:\
Acudimos ahora al siguiente ejemplo: $f(x,y)=x^{2n}+y^{2m}$, $m,n\in\mathbb{N}_{>1}$, $\Omega=\mathbb{R}^2$, y $x_*=(0,0)$\
Calculamos sus derivadas parciales de primer y segundo orden:\
Como $x^{2n}\geq 0$ $\forall x\in\mathbb{R}$ e $y^{2m}\geq 0$ $\forall y\in\mathbb{R}$, tenemos que $x^{2n}+y^{2m}\geq 0$\
Al evualuar $f$ en nuestro $x_*$ observamos que $f(0,0)=0$ y así $x_*$ es un minimizador local.\
Calculamos las derivadas parciales de primer y segundo orden de la función objetivo $f$:\
$f_x(x,y)=2nx^{2n-1}$; $f_y(x,y)=2my^{2m-1}$\
$f_{xx}(x,y)=2n(2n-1)x^{2n-2}$; $f_{xy}(x,y)=0=f_{yx}(x,y)$; $f_{yy}(x,y)=2m(2m-1)y^{2m-2}$\
Luego, armamos su matriz Hessiana:\
$$
\left( \begin{array}{lll}
2n(2n-1)x^{2n-2} & 0\\ 0 & 2m(2m-1)y^{2m-2}
\end{array}\right)
$$ 
Que se anula si $x=y=0$, por lo tanto no está definida positiva por ser la transformación nula. ($x^T 0_{2 \times 2} x= 0$ $\forall x \in \mathbb{R}^2$)
 

$OP-P1-E20$\
Queremos minimizar $f:\mathbb{R}^n\rightarrow\mathbb{R}$ dada por $f(x)=c^Tx$ sujeto a las condiciones:\
$(1)$ $\sum\limits_{i=1}^nx_i=0$, y $(2)$  $\sum\limits_{i=1}^nx_i^2=1$\
Como $c\in\mathbb{R}^n$, entonces $c=(c_1,...,c_n)$, y así $f(x)=\sum\limits_{i=1}^nc_ix_i$ es suave (en particular diferenciable, que es lo que nos interesa)con $\nabla f(x)=(c_1,...,c_n)$, $g(x)=\sum\limits_{i=1}^nx_i$ y $h(x)=\sum\limits_{i=1}^nx_i^2$ son diferenciables y tales que:\
$\nabla g(x)=(1,..,1)\neq 0$ y $\nabla h(x)=(2x_1,..,2x_n)\neq 0$ en los dominios de interés. Por lo tanto, estamos ante las hipótesis necesarias para aplicar el Teroema de los Multiplicadores de Lagrange, luego existen $\lambda$ y $\mu$ tales que

$\nabla f(x)=\lambda\nabla g(x)+\mu\nabla h(x)$, es decir\
$(c_1,...,c_n)=\lambda(1,..,1)+2\mu(x_1,..,x_n)$ de aquí que:\
$f_{x_i}=c_i=\lambda+2\mu x_i$,$(3)$ y así:

$\sum\limits_{i=1}^nc_i=\sum\limits_{i=1}^n(\lambda+2\mu x_i)=$$n\lambda+2\mu\sum\limits_{i=1}^nx_i=n\lambda$ por $(1)$. Obteniendo:

$\lambda=\frac{\sum\limits_{i=1}^nc_i}{n}$.
Ahora, por $(3)$,

$\frac{c_i-\lambda}{2\mu}=x_i$ $(4)$ $\Rightarrow $ $\left(\frac{c_i-\lambda}{2\mu}\right)^2= x_i^2$ $\Rightarrow  \sum\limits_{i=1}^n $$\left(\frac{c_i-\lambda}{2\mu}\right)^2=\sum\limits_{i=1}^nx_i^2=1\Leftrightarrow \mu^2=\sum\limits_{i=1}^n\left (\frac{c_i-\lambda}{4}\right)^2$
$\Leftrightarrow \mu=\pm\sqrt{\sum\limits_{i=1}^n\left (\frac{c_i-\lambda}{4}\right)^2}$

por $(4)$ $x_i= \frac{c_i-\frac{\sum\limits_{i=1}^nc_i}{n}}{2\mu}$=$\pm \frac{c_i-\frac{\sum\limits_{i=1}^nc_i}{n}}{2\cdot\sqrt{\sum\limits_{i=1}^n\left(\frac{c_i-\frac{\sum\limits_{i=1}^nc_i}{n}}{4}\right)^2}}$ $=\pm \frac{c_i-\frac{\sum\limits_{i=1}^nc_i}{n}}{\sqrt{\sum\limits_{i=1}^n\left (c_i-\frac{\sum\limits_{i=1}^nc_i}{n}\right)^2}}$. 

Entonces:

$x_+= \left(\frac{c_1-\frac{\sum\limits_{i=1}^nc_i}{n}}{\sqrt{\sum\limits_{i=1}^n\left (c_i-\frac{\sum\limits_{i=1}^nc_i}{n}\right)^2}},..., \frac{c_n-\frac{\sum\limits_{i=1}^nc_i}{n}}{\sqrt{\sum\limits_{i=1}^n\left (c_i-\frac{\sum\limits_{i=1}^nc_i}{n}\right)^2}}\right)$ y;

$x_-= \left(-\frac{c_1-\frac{\sum\limits_{i=1}^nc_i}{n}}{\sqrt{\sum\limits_{i=1}^n\left (c_i-\frac{\sum\limits_{i=1}^nc_i}{n}\right)^2}},..., -\frac{c_n-\frac{\sum\limits_{i=1}^nc_i}{n}}{\sqrt{\sum\limits_{i=1}^n\left (c_i-\frac{\sum\limits_{i=1}^nc_i}{n}\right)^2}}\right)$

Evaluando, obtenemos:

$f(x_+)= \sum\limits_{i=1}^n  \frac{c_i^2-c_i\frac{\sum\limits_{i=1}^nc_i}{n}}{\sqrt{\sum\limits_{i=1}^n\left (c_i-\frac{\sum\limits_{i=1}^nc_i}{n}\right)^2}}$ y, $f(x_-)=-\sum\limits_{i=1}^n  \frac{c_i^2-c_i\frac{\sum\limits_{i=1}^nc_i}{n}}{\sqrt{\sum\limits_{i=1}^n\left (c_i-\frac{\sum\limits_{i=1}^nc_i}{n}\right)^2}}$.

Ahora bien:\
si $\sum\limits_{i=1}^n c_i^2>\sum\limits_{i=1}^nc_i\frac{\sum\limits_{i=1}^nc_i}{n}$, $f(x_+)>0>f(x_-)$ entonces  $x_-$ es el mínimo buscado.\
si $\sum\limits_{i=1}^n c_i^2=\sum\limits_{i=1}^nc_i\frac{\sum\limits_{i=1}^nc_i}{n}$, entonces $c_i=\lambda$ $\forall i=1:n$ y $f(x)=0$ $\forall x\in\mathbb{R}^n$ pues $f(x)=\sum\limits_{i=1}^nc_ix_i=\lambda\sum\limits_{i=1}^nx_i=\lambda\cdot 0=0$; y\
si $\sum\limits_{i=1}^n c_i^2<\sum\limits_{i=1}^nc_i\frac{\sum\limits_{i=1}^nc_i}{n}$, $f(x_-)>0>f(x_+)$ entonces $x_+$ es el mínimo buscado.

$OP-P2-E4$\
Veamos que dadas $g:\mathbb{R}\rightarrow\mathbb{R}$ no decreciente y convexa en $\mathbb{R}$ y $f:\mathbb{R}^n\rightarrow\mathbb{R}$ convexa en $\Omega\subset\mathbb{R}^n$, $g\circ f$ es convexa en $\Omega$.

Si una función $h$ es convexa en un $D\subset Domh$ , tenemos que $h(t\cdot x+(1-t)\cdot y)\leq t\cdot h(x)+(1-t)\cdot h(y)$ ,  $\forall t\in[0,1]$,  $\forall x,y \in D$ $(*)$.

Dado que $g$ es no decreciente, si $a\leq b$, $g(a)\leq g(b)$, $\forall a,b \in Domg$  $(**)$. Luego:

$g\circ f(t\cdot x+(1-t)\cdot y)= g(f(t\cdot x+(1-t)\cdot y))\leq g(t\cdot f(x)+(1-t)\cdot f(y))$   $\forall t\in[0,1]$,  $\forall x,y \in\Omega $ Por $(*)$ y $(**)$ pues f es convexa en $\Omega$ y $(**)$; y

$g(t\cdot f(x)+(1-t)\cdot f(y))\leq t\cdot g(f(x))+(1-t)\cdot g(f(x))$ por $(*)$.

Por lo tanto, $g\circ f$ es convexa.

$OP-P2-E8$\
Tenemos $f:\mathbb{R}^n\rightarrow\mathbb{R}$ de clase $C^1$ y para $k\in\mathbb{N}$ la sucesión dada por $x^{k+1}=x^k-\lambda_k\nabla f(x^k)$. Con $\lambda_k\geq\bar{\lambda}>0$\
Si {${x^k}$}$_{k=0}^{\infty}$ tiende a $x^*$, veamos que $\nabla f(x^*)=0$.

Que $x^k\rightarrow x^*$ significa que $\forall\varepsilon >0$,  $\exists N\in\mathbb{N}$ tal que si $n\geq N$ entonces $|| x^n- x^*||<\varepsilon$ 

Como estamos en un Banach las sucesiones convergen si y sólo si son de Cauchy. Luego, lo anterior equivale a decir que:

$\forall\varepsilon >0$,  $\exists N\in\mathbb{N}$ tal que si $m,n\geq N$ entonces $|| x^n- x^m||<\varepsilon$ 

Sin pérdida de generalidad, podemos suponer que $m<n$ y tomamos $n=m+1$. Así:

 $|| x^n- x^m|| =|| x^{m+1}- x^m||=||x^m-\lambda_m\nabla f(x^m)- x^m||=||\lambda_m\nabla f(x^m)||<\varepsilon$. Luego:
 
 $||\lambda_m\nabla f(x^m)||=|\lambda_m|||\nabla f(x^m)||=\lambda_m||\nabla f(x^m)||<\varepsilon\Leftrightarrow ||\nabla f(x^m)||<\frac{\varepsilon}{\lambda_m}\leq\frac{\varepsilon}{\bar{\lambda}}$ para $m$ suficientemente grande.
 
 Como $f$ es de clase $C^1$, su gradiente, $\nabla f$, es una función continua, y así:\
 $\nabla f(x^*)=\lim\limits_{k\rightarrow\infty}\nabla f(x^k)=0$ por lo visto arriba tomando $\delta=\bar{\lambda}\varepsilon$ en la definición de límite. 
 
 Es decir $\nabla f(x^*)=0$ como queríamos.
